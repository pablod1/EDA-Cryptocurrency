# Pandas csv formatting

```python
df = pd.read_csv('../../data/crypto_csv_test.csv')
```

### 1. Price

```python
# Replace all special characters and symbols
df['price'] = df['price'].str.replace('$','')
df['price'] = df['price'].str.replace('<','')
df['price'] = df['price'].str.replace(',','')
df['price'] = df['price'].str.replace('--','0')

# Convert column to float
df['price'] = df['price'].astype('float')
```

```python
df.tail()
```

### 2. Mcap

```python
# Replace all special characters and symbols
df['mcap'] = df['mcap'].str.replace('$','')
df['mcap'] = df['mcap'].str.replace('--','0')
df['mcap'] = df['mcap'].str.replace(',','')

# Convert column to float
df['mcap'] = df['mcap'].astype('float')
```

```python
df.tail()
```

### 3. Circulating supply

```python
# Remove letters
df['circulating_supply'] = df['circulating_supply'].str.replace('\D','')

# Replace empty values with 0
df.loc[df['circulating_supply'] == '','circulating_supply'] = '0'

# Remove values with 20 or more chars (only 3 values)
df = df[df['circulating_supply'].str.count('[0-9]') < 20]

# Convert to int64
df['circulating_supply'] = df['circulating_supply'].astype('int64')
```

```python
df.tail()
```

### 4. Date

```python
# Create 2 columns more: year and month
df['year'] = df['date'].str[:4]
df['month'] = df['date'].str[5:7]

# Convert year and month column to int
df['year'] = df['year'].astype('int32')
df['month'] = df['month'].astype('int32')

# Convert to datetime object
df['date'] = pd.to_datetime(df['date'])
```

```python
df.tail()
```

### Convert the dataframe into csv but with formatted data

```python
df.to_csv('../../data/crypto_csv_formatted_test.csv',index=False)
```
